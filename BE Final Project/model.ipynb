```python
image_list= []
resized_image= []
file_list = []
```

```python
for filename in glob.glob('C:/Users/Prasad/croping/a/*.jpg'):
    print(filename)
    file_list.append(filename)
    img= Image.open(filename)
    image_list.append(img)
```

```python
for image in image_list:
    #image.show()
   # image= image.resize((100,100))
    # resized_image.append(image)
    ########################
    
    
   
    image= image.crop((150,150,300,300))
    resized_image.append(image)
    
    
    print(image)
print(len(resized_image))
```

```python

for (i,new) in enumerate(resized_image):
    print(file_list)
    print(i)
    file=ntpath.basename(file_list[i])
    #print(file)
    new.save('{}{}'.format('C:/Users/Prasad/croping_paste/',file))
   # new.show()
   

   
```
